```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("/content/drive/MyDrive/College_data_updated.csv", encoding='ISO-8859-1')


data.replace('--', np.nan, inplace=True)


label_encoder = LabelEncoder()
data['Faculty'] = label_encoder.fit_transform(data['Faculty'])
data['Social_Life'] = label_encoder.fit_transform(data['Social_Life'])


features = ['Rating', 'Academic', 'Faculty', 'Infrastructure', 'Placement', 'Social_Life']
target = 'Placement_Rate'


train_data = data.iloc[:20]
test_data = data.iloc[20:]

X_train = train_data[features]
y_train = train_data[target]
X_test = test_data[features]


imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
y_train_imputed = imputer.fit_transform(y_train.values.reshape(-1, 1)).ravel()
scaler = StandardScaler()

# Fit and transform the scaler on the training data
X_train_scaled = scaler.fit_transform(X_train_imputed)

# Apply the same transformation to the test data
X_test_scaled = scaler.transform(X_test_imputed)


model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train_imputed)


predictions = model.predict(X_test_scaled)


mse = mean_squared_error(y_train_imputed, model.predict(X_train_scaled))
rmse = np.sqrt(mse)
print("MSE:", mse)
print("RMSE:", rmse)


y_train_pred = model.predict(X_train_scaled)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_train_imputed, y=y_train_pred)
plt.xlabel('Actual Placement Rate')
plt.ylabel('Predicted Placement Rate')
plt.title('Actual vs Predicted Placement Rates (Training Set)')
plt.show()
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load Data
data = pd.read_csv("/content/drive/MyDrive/College_data_updated.csv", encoding='ISO-8859-1')

# Replace '--' or invalid entries with NaN
data.replace('--', pd.NA, inplace=True)

# Handle Missing Values (Convert to numeric and fill NaN)
data['PG_fee'] = pd.to_numeric(data['PG_fee'], errors='coerce')
data['PG_fee'] = data['PG_fee'].fillna(data['PG_fee'].mean())

data['UG_fee'] = pd.to_numeric(data['UG_fee'], errors='coerce')
data['UG_fee'] = data['UG_fee'].fillna(data['UG_fee'].mean())

# Encode Categorical Data
encoder = LabelEncoder()
data['Stream'] = encoder.fit_transform(data['Stream'])
data['State'] = encoder.fit_transform(data['State'])

# Feature Selection
features = ['Stream', 'Rating', 'Faculty', 'UG_fee', 'PG_fee', 'Placement']
X = data[features]

# Ensure all columns are numeric and handle NaN explicitly
X = X.apply(pd.to_numeric, errors='coerce')
X.fillna(0, inplace=True)

# Normalize Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KNN Model
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_scaled, data['College_Name'])
user_input = [[0, 9.0, 8.3, 600000, 50000, 8.5]]
scaled_input = scaler.transform(user_input)


recommendations = model.kneighbors(scaled_input, n_neighbors=5, return_distance=False)
print("Recommended Colleges:")
for rec in recommendations[0]:
    print(data.iloc[rec]['College_Name'])
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Load Existing Dataset
data = pd.read_csv("/content/drive/MyDrive/College_data_updated.csv", encoding='ISO-8859-1')

# Function to Fetch Course Data
def fetch_courses(college_name):
    try:
        query = '+'.join(college_name.split())  # Format college name for search query
        url = f"https://www.shiksha.com/search?q={query}"  # Replace with actual target website
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract course info (modify selectors as per website structure)
        courses = []
        for course in soup.find_all('div', class_='course-name'):
            courses.append(course.text.strip())

        return ', '.join(courses[:5])  # Limit to top 5 courses
    except:
        return "Not Available"

# Add Courses to Dataset
data['Courses_Offered'] = data['College_Name'].apply(fetch_courses)

# Save Updated Dataset
data.to_csv("/content/drive/MyDrive/College_data_updated.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data (larger dataset)
data = pd.DataFrame({
    'College_Name': ['College A', 'College B', 'College C', 'College D', 'College E', 'College F', 'College G'],
    'Rating': [4, 3, 5, 2, 4, 3, 5],
    'Academic': [5, 4, 5, 3, 4, 3, 5],
    'Faculty': [3, 3, 5, 2, 4, 3, 5],
    'Infrastructure': [4, 3, 5, 2, 4, 3, 5],
    'Placement': [4, 3, 5, 2, 4, 3, 5],
    'Social_Life': [5, 4, 5, 3, 4, 3, 5],
    'Placement_Rate': [90.0, 70.0, 102.0, 50.0, 82.0, np.nan, np.nan]  # Missing Placement Rate for College F and G
})

# Features and target
features = ['Rating', 'Academic', 'Faculty', 'Infrastructure', 'Placement', 'Social_Life']
target = 'Placement_Rate'

# Separate data into known and missing placement rates
data_known = data.dropna(subset=[target])  # Known Placement Rate (College A, B, C, D, E)
data_missing = data[data[target].isnull()]  # Missing Placement Rate (College F, G)

# Train the model on known data (College A, B, C, D, E)
X_known = data_known[features]
y_known = data_known[target]

# Impute missing values in the features (if any)
imputer = SimpleImputer(strategy='mean')
X_known_imputed = imputer.fit_transform(X_known)

# Scale the features
scaler = StandardScaler()
X_known_scaled = scaler.fit_transform(X_known_imputed)

# Train a Linear Regression model
model = LinearRegression()
model.fit(X_known_scaled, y_known)

# Predict Placement Rate for missing data (College F, G)
X_missing = data_missing[features]
X_missing_imputed = imputer.transform(X_missing)
X_missing_scaled = scaler.transform(X_missing_imputed)

# Make predictions for the missing Placement Rates
predictions = model.predict(X_missing_scaled)

# Assign the predicted values to the missing data
data.loc[data[target].isnull(), 'Placement_Rate'] = predictions

# Show the updated dataset
print(data)

# Evaluate the model with MSE and RMSE
y_known_pred = model.predict(X_known_scaled)
mse = mean_squared_error(y_known, y_known_pred)
rmse = np.sqrt(mse)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)

# Visualize Actual vs Predicted Placement Rates (for the training data)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_known, y=y_known_pred)
plt.xlabel('Actual Placement Rate')
plt.ylabel('Predicted Placement Rate')
plt.title('Actual vs Predicted Placement Rates (Training Data)')
plt.show()
```
